Question 5) Create a new column named as 'happiness_degree' which labels value of 'ladder'
*very happy if ladder < 30
*happy if 30 <= ladder < 60
*moderately happy if 60 <= ladder < 90
*unhappy if 90 <= ladder < 120
*very unhappy if ladder >= 120
Print the top 10 records updated dataframe (10 points)

```python
#codes
conditions = [
  (happiness_df['Ladder']  < 30),
  (happiness_df['Ladder']  < 60) & (happiness_df['Ladder']  >= 30),
  (happiness_df['Ladder']  < 90) & (happiness_df['Ladder']  >= 60),
  (happiness_df['Ladder']  < 120) & (happiness_df['Ladder']  >= 90),
  (happiness_df['Ladder']  >= 120)
]

values = ['very happy', 'happy', 'moderately happy', 'unhappy', 'very unhappy']

happiness_df['happiness_degree'] = np.select(conditions, values)
happiness_df.head(10)
```

```python
#codes
df = happiness_df.iloc[:,1:11]
df.head()
happiness_df.iloc[0, 1:]
```

```python
#codes
cor_mat = happiness_df[['Ladder', 'SD of Ladder', 'Positive affect', 'Negative affect', 'Social support', 'Freedom', 'Corruption', 'Generosity', 'Log of GDP\nper capita', 'Healthy life\nexpectancy']].corr()
mask = np.array(cor_mat)
mask[:] = False
fig = plt.gcf()
fig.set_size_inches(30,12)
sns.heatmap(data=cor_mat,mask=mask,vmax=1.0, square=True,annot=True,cbar=True)
```

- here is just the definition of a heat map if you want my comment you should read below.

Each square shows the correlation between the variables on each axis. Correlation ranges from -1 to +1. Values closer to zero means there is no linear trend between the two variables. The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar, but instead of both increasing one variable will decrease as the other increases. The diagonals are all 1/white because those squares are correlating each variable to itself (so it's a perfect correlation). For the rest the larger the number and lighter the color the higher the correlation between the two variables. The plot is also symmetrical about the diagonal since the same two variables are being paired together in those squares.

By definition and heat map, you can see that our DataFrame have weak linear correlation for most variable.But some of them have strong linear correlated. As you can see healhy life expectancy has strong correlation with Ladder, SD of Ladder, Social support, Log of GDP per capital. It reason probably they are about a good quality of life.And ıt has weak correlation with corruption. Log of GDP per capital has strong correlation with social support and and ladder. That means if social support increase, Log of GDP per capital will increase. other correlation values ​​are weaker than what I wrote before. Question 9) Plot multiple bar charts in a single chart that show the mean values of 'Social support','Freedom','Generosity' according to happiness degree (15 points)

```python
a = df['Social support'].groupby(by = 'happiness_degree').mean()
b = df['Freedom'].groupby(by = 'happiness_degree').mean()
c = df['Generosity'].groupby(by = 'happiness_degree').mean()
frames = pd.concat([a, b, c], axis=1)
ax =frames.plot(kind='bar',xticks=np.arange(5), stacked=False, figsize= (10, 10))
ax.set_xticklabels(["happy", "moderately happy", "unhappy", "very happy", "very unhappy"])
plt.xlabel("Happiness Score")
plt.title('Happiness Score of the Countries')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
